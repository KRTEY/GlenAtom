```python
# In[1]:
import glenatom as Glen
import matplotlib.pyplot as plt
import numpy as np

print(f"GlenAtom library imported successfully!")
```

```python
# In[2]:
styles = Glen.get_available_styles()
print(f"✅ Available styles: {styles}")
```

```python
# In[3]:
# 'line' 스타일 적용
Glen.fiddich(what='line')

# 그래프에 쓰일 데이터 생성
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x) * np.exp(-0.2 * x)

# 그래프 그리기
fig, ax = plt.subplots(figsize=(6, 4)) # figsize은 코드에서 최종 조절 가능
ax.plot(x, y, label='Damped Sine Wave')
ax.set_title("GlenAtom 'line' Style Example")
ax.set_xlabel("Time (s)")
ax.set_ylabel(r"Signal Strength ($V$)") # LaTex 문법도 잘 적용됩니다.
ax.legend()
ax.grid(True) # 스타일 시트의 grid 설정을 따릅니다.

plt.show()
```

```python
# In[4]:
# 'bar' 스타일로 변경
Glen.fiddich(what='bar')

# 막대그래프에 쓰일 데이터
categories = ['Group A', 'Group B', 'Group C', 'Group D']
values = [15, 28, 22, 35]

# 그래프 그리기
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(categories, values, color='skyblue', edgecolor='black')
ax.set_title("GlenAtom 'bar' Style Example")
ax.set_xlabel("Category")
ax.set_ylabel("Count")

plt.show()
```
